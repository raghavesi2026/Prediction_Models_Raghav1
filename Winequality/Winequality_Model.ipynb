```python

import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

# Read CSV File
df3 = pd.read_csv('/content/winequality.csv')

# Split Data Into Input And Output Variable
x = df3.drop(['quality'],axis = 1)
y = df3['quality']

# Train Test Split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 42)

# Model Create
model = RandomForestRegressor(n_estimators=1000,max_depth=15,random_state=42)

# Model Train
model.fit(xtrain,ytrain)

# Model Predict
ypred = model.predict(xtest)

# Find Accurcay
accu = r2_score(ytest,ypred)
print("Accuracy :- ",accu)
```
